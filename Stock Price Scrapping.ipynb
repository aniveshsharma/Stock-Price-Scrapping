```python
#importing Dependencies

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# url from where data is Scrape
url = "https://www.investing.com/equities"
```

```python
# Sending a Request to Website to See the website is Scrapeable or not.
page = requests.get(url)

print(page)
```

```python
# After Getting Response 200 we are now start the process of Scrapping

soup = BeautifulSoup(page.text, "html")

print(soup.prettify())
```

```python
#Now we find the id or Class name of the Stocks

stocks = soup.find("tbody", class_= "datatable-v2_body__8TXQk")
print(stocks.text)
```

```python
stock = stocks.find_all("tr", class_="datatable-v2_row__hkEus dynamic-table-v2_row__ILVMx")
for stock in stocks:
    print(stock.text)
```

```python
stockes = {
    "Company": [],
    "Last": [],
    "High": [],
    "Low": [],
    "Change": [],
    "Change(%)": [],
    "volumn": [],
    "Time": []
}

stock = soup.find("tbody", class_="datatable-v2_body__8TXQk")
stocks = stock.find_all('tr', class_='datatable-v2_row__hkEus dynamic-table-v2_row__ILVMx')

for i in stocks:
    company = i.find("td", class_='datatable-v2_cell__IwP1U !h-auto w-full py-2 mdMax:border-r dynamic-table-v2_col-name__Xhsxv !py-2').text.strip()
    
    last = i.find_all('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right')[0].text.strip()
    high = i.find_all('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right')[1].text.strip()
    low = i.find_all('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right')[2].text.strip()

    change = i.find('td', class_='datatable-v2_cell__IwP1U datatable-v2_cell--up__lVyET datatable-v2_cell--bold__cXQUV dynamic-table_col-other__Eu_RC text-right font-bold rtl:text-right')
    if not change:
        change = i.find('td', class_='datatable-v2_cell__IwP1U datatable-v2_cell--down__sYmZ4 datatable-v2_cell--bold__cXQUV dynamic-table_col-other__Eu_RC text-right font-bold rtl:text-right')
    change = change.text.strip()

    change_pct = i.find('td', class_='datatable-v2_cell__IwP1U datatable-v2_cell--up__lVyET datatable-v2_cell--bold__cXQUV dynamic-table-v2_col-other__zNU4A text-right font-bold rtl:text-right')
    if not change_pct:
        change_pct = i.find('td', class_='datatable-v2_cell__IwP1U datatable-v2_cell--down__sYmZ4 datatable-v2_cell--bold__cXQUV dynamic-table-v2_col-other__zNU4A text-right font-bold rtl:text-right')
    change_pct = change_pct.text.strip()

    volumn = i.find('td', {'data-test': 'volume-cell'}).text.strip()
    time = i.find('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A').text.strip()

    stockes['Company'].append(company)
    stockes['Last'].append(last)
    stockes['High'].append(high)
    stockes['Low'].append(low)
    stockes['Change'].append(change)
    stockes['Change(%)'].append(change_pct)
    stockes['volumn'].append(volumn)
    stockes['Time'].append(time)
```

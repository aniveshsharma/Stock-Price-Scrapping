```python
#importing Dependencies

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# url from where data is Scrape
url = "https://www.investing.com/equities"
```

```python
# Sending a Request to Website to See the website is Scrapeable or not.
page = requests.get(url)

print(page)
```

```python
# After Getting Response 200 we are now start the process of Scrapping

soup = BeautifulSoup(page.text, "html")

print(soup.prettify())
```

```python
#Now we find the id or Class name of the Stocks

stocks = soup.find("tbody", class_= "datatable-v2_body__8TXQk")
print(stocks.text)
```

```python
stock = stocks.find_all("tr", class_="datatable-v2_row__hkEus dynamic-table-v2_row__ILVMx")
for stock in stocks:
    print(stock.text)
```

```python
stockes = {
    "Company": [],
    "Last": [],
    "High": [],
    "Low": [],
    "Change": [],
    "Change(%)": [],
    "volumn": [],
    "Time": []
}

stock = soup.find("tbody", class_="datatable-v2_body__8TXQk")
stocks = stock.find_all('tr', class_='datatable-v2_row__hkEus dynamic-table-v2_row__ILVMx')

for i in stocks:
    company = i.find("td", class_='datatable-v2_cell__IwP1U !h-auto w-full py-2 mdMax:border-r dynamic-table-v2_col-name__Xhsxv !py-2').text
    last = i.find_all('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right')[0].text
    high = i.find_all('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right')[1].text
    low = i.find_all('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right')[2].text
    
    # Handling both positive and negative changes
    change = i.find('td', class_=['datatable-v2_cell--up__lVyET', 'datatable-v2_cell--down__sYmZ4'])
    change_pct = i.find('td', class_=['datatable-v2_cell--up__lVyET', 'datatable-v2_cell--down__sYmZ4'])

    # Extracting text or setting to 'N/A' if not found
    change = change.text if change else 'N/A'
    change_pct = change_pct.text if change_pct else 'N/A'
    
    volumn = i.find('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A text-right rtl:text-right').text
    time = i.find('td', class_='datatable-v2_cell__IwP1U dynamic-table-v2_col-other__zNU4A').text

    stockes['Company'].append(company)
    stockes['Last'].append(last)
    stockes['High'].append(high)
    stockes['Low'].append(low)
    stockes['Change'].append(change)
    stockes['Change(%)'].append(change_pct)
    stockes['volumn'].append(volumn)
    stockes['Time'].append(time)
```
